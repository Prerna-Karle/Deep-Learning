```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense
```

#1. Prepare Dataset

```python
# Input sequences
X = np.array([
[10, 20, 30],
[20, 30, 40],
[30, 40, 50],
[40, 50, 60]
])
# Output values
y = np.array([40, 50, 60, 70])
```

```python
# Reshape input to 3D for RNN
# Shape required: (samples, timesteps, features)
X = X.reshape((X.shape[0], X.shape[1], 1))
```

#2. Build Simple RNN Model

```python
model = Sequential()
# Simple RNN layer
model.add(SimpleRNN(units=50, activation='relu', input_shape=(3, 1)))
# Output layer
model.add(Dense(1))
# Compile model
model.compile(optimizer='adam', loss='mse')
```

#3. Train Model

```python
model.fit(X, y, epochs=500, verbose=0)
```

#4. Test Prediction

```python
test_input = np.array([[50, 60, 70]])
test_input = test_input.reshape((1, 3, 1))
prediction = model.predict(test_input)
print("Predicted Output:", prediction[0][0])
```
